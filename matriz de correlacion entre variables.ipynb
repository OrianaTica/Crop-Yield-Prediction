```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_rendimiento=pd.read_csv('Df_rendimiento.csv',index_col=0)
df_rendimiento.head()
```

```python
# Calcular matriz de correlación
columnas_seleccionas=['hg/ha_yield','average_rain_fall_mm_per_year','pesticides_tonnes','avg_temp']
df_seleccionado = df_rendimiento[columnas_seleccionas]
correlation_matrix = df_seleccionado.corr()

# Crear heatmap de correlación
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title('Matriz de Correlación entre las Variables')
plt.show()
```
